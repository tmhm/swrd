## data1 analysis

The avg reward: 580.992688163

The avg learn time: 181.230236723 s,

```python
from numpy import mean
data_filename="./data/data1.csv"
pf=open(data_filename, 'r')
cnt = 0
reward_list = []
for line in pf.readlines():
    #d=line.split('test reward: ')
    d=line.split('\r') 
    #print line
    #print d[0]
    for dx in d:
        if str(dx).startswith('test reward: '):
            cnt +=1
            reward_list.append(float(dx[12:-1]))
pf.close()
print cnt
#print reward_list   
rmean=mean(reward_list)
print rmean
re_cnt = {'cnt1': 0, 'cnt2':0}

# cnt list for all reward
Cnt_list1 = []
Cnt_list2 = []

# count number
cnt1_t = 0
cnt2_t = 0
for re in reward_list:
    if re > 2000:
        cnt2_t +=1
        re_cnt['cnt2'] +=1
    Cnt_list2.append(cnt2_t)

            
    if re > 1000:
        cnt1_t +=1
        re_cnt['cnt1'] +=1
    Cnt_list1.append(cnt1_t)
    
print re_cnt
plt.subplot(1, 2, 1)
#plt.figure(1)
plt.plot(reward_list)
#plt.figure(2)
plt.subplot(1, 2, 2)
plt.plot(Cnt_list1)
plt.plot(Cnt_list2)

#if write_flag is True:
#    pf = open(reward_write_finename,'a')
#    pf.write(str(Cnt_list1)[1:-1]+ '\n')
#    pf.close()
reward_dict['rew1'] = Cnt_list1
Cntreward_dict['cnt1'] = Cnt_list2
```

## data3 analysis


The avg reward: 820.104630293

The avg learn time: 1067.88101318 s on data[ 2016-07-07 ],

```python
from numpy import mean
data_filename="./data/data3.csv"
pf=open(data_filename, 'r')
cnt = 0
reward_list = []
for line in pf.readlines():
    #d=line.split('test reward: ')
    d=line.split('\r') 
    #print line
    #print d[0]
    for dx in d:
        #if str(dx).startswith('test reward: '):
        #    cnt +=1
        #    reward_list.append(float(dx[12:-1]))
        if len(reward_list) == 300:
            break
        rew=dx.split(':')
        #print rew[1]
        reward_list.append(float(rew[1]))      
pf.close()
#print cnt
#print reward_list   
rmean=mean(reward_list)
print rmean
re_cnt = {'cnt1': 0, 'cnt2':0}

# cnt list for all reward
Cnt_list1 = []
Cnt_list2 = []

# count number
cnt1_t = 0
cnt2_t = 0
for re in reward_list:
    if re > 2000:
        cnt2_t +=1
        re_cnt['cnt2'] +=1
    Cnt_list2.append(cnt2_t)

            
    if re > 1000:
        cnt1_t +=1
        re_cnt['cnt1'] +=1
    Cnt_list1.append(cnt1_t)
    
print re_cnt
plt.subplot(1, 2, 1)
#plt.figure(1)
plt.plot(reward_list)
#plt.figure(2)
plt.subplot(1, 2, 2)
plt.plot(Cnt_list1)
plt.plot(Cnt_list2)

reward_dict['rew3'] = Cnt_list1
Cntreward_dict['cnt3'] = Cnt_list2
```
